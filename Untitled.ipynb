```python
import pandas as pd
```

```python
df = pd.read_json('data/export.geojson')
```

```python
df_features = df.features.apply(pd.Series)
```

```python
(df_features).head()
```

```python
df_geom = df_features.geometry.apply(pd.Series)
```

```python
df_geom.head()
```

```python
df_properties = df_features.properties.apply(pd.Series)
```

```python
df_properties.head()
```

```python
df_geom_properties = pd.concat([df_properties,df_geom],axis=1)
```

```python
df_geom_properties = df_geom_properties[df_geom.type=='Point']
```

```python
df_coord = pd.DataFrame(df_geom_properties.coordinates.tolist(), columns=['lon', 'lat'])
```

```python
df_geom_properties = pd.concat([df_geom_properties['name'],df_coord],axis=1)
```

```python
df_geom_properties.to_csv('data/bvg_stations.csv')
```
